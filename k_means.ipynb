## np.random.seed()

Computers follow a set of instructions to do anything; to generate random numbers as well; they use some seed. Hence random numbers are not truly random, they are hence called pseudorandom numbers.

np.random.seed() makes the random number predictable.

```python
np.random.seed(0)
np.random.choice(10, size=10)
```

## Data generation for K-means

```python
np.random.seed(0)
N = 2000
data1 = np.random.randn(N//3, 2) + np.array([2,2])
data2 = np.random.randn(N//3, 2) + np.array([-3,-3])
data3 = np.random.randn(N//3, 2) + np.array([-5,3])
# X = np.concatenate((data1, data2, data3), axis=0)
X = np.concatenate((data1, data2, data3))
X.shape
```

## Visualize data

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1])
```

## Initialize cluster centers

```python
np.random.seed(0)
k = 3
random_indices = np.random.randint(0, X.shape[0], size=k)
print(random_indices)
cluster_centers = X[random_indices]
print(cluster_centers)
```

## Visualize the cluster centers as well

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(cluster_centers[:,0], cluster_centers[:, 1], marker='+', color='black', s=200);
```

## Assign data points to these three clusters (store this information in the cluster indicator (z)) 

<!-- Compute cluster indicator (z) -->

```python
z = np.array([])
for point in X:
  dist_from_each_center = [np.linalg.norm(point - cluster_centers[i]) for i in range(k)]
  choosen_cluster = np.argmin(dist_from_each_center)
  z = np.append(z, choosen_cluster)
z, np.unique(z)
```

```python
def identify_cluster_for_each_data_point(X, k, cluster_centers):
  z = np.array([])
  # z = []
  for point in X:
    dist_from_each_center = [np.linalg.norm(point - cluster_centers[i]) for i in range(k)]
    choosen_cluster = np.argmin(dist_from_each_center)
    # print(choosen_cluster)
    z = np.append(z, choosen_cluster)
    # z.append(choosen_cluster)
  return z
```

```python
z = identify_cluster_for_each_data_point(X, k, cluster_centers)
np.unique(z)
```

## Visualize the clusters formed as per this z.

```python
plt.scatter(X[:,0], X[:,1], c=z)
plt.scatter(cluster_centers[:,0], cluster_centers[:, 1], marker='+', color='k')
```

## Compute updated cluster centers

```python
def compute_cluster_centers(X, k, z):
  cluster_centers = np.array([np.mean(X[z==i], axis=0) for i in range(k)])
  return cluster_centers
```

```python
cluster_centers = compute_cluster_centers(X, k, z)
```

```python
plt.scatter(X[:,0], X[:,1], c=z)
plt.scatter(cluster_centers[:,0], cluster_centers[:, 1], marker='+', color='red')
```

## Iterations involving 

(a) re-assigning data points to new clusters as per the new cluster centers formed.

(b) Re-compute cluster centers

```python
for i in range(5):
  z = identify_cluster_for_each_data_point(X, k, cluster_centers)
  cluster_centers = compute_cluster_centers(X, k, z)
  print(cluster_centers)
  plt.scatter(X[:,0], X[:,1], c=z)
  plt.scatter(cluster_centers[:,0], cluster_centers[:, 1], marker='+', color='red')
  plt.show()
```
